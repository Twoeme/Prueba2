```python
#Ejercicio 3: Búsqueda y Reporte de Anomalías
#Simular una lista de mediciones técnicas donde algunos elementos sean claramente errores (valores negativos, o números muy elevados). Escribir un algoritmo que recorra la lista, detecte las anomalías y genere un reporte indicando la posición y el valor del dato anómalo.

# Lista con mediciones donde hay errores como valores negativos o muy altos
mediciones = [10, 15, -3, 20, 1000, 18, 25, -1]

print("Reporte de anomalías:")

for i in range(len(mediciones)):
    valor = mediciones[i]
    if valor < 0 or valor > 100:
        print(f"Anomalía en posición {i}: valor {valor}")
```

```python
#Ejercicio 4: Generación de Informes Semanales
#Con una lista de datos diarios (por ejemplo, humedad del suelo durante 7 días), pedir que calcule día a día y semana a semana: máximos, mínimos y promedio. El informe se presenta como texto, estructurado con títulos para cada parámetro.

# Datos de humedad diaria en porcentaje durante una semana
humedad_dias = [55, 60, 58, 57, 62, 59, 61]

max_humedad = humedad_dias[0]
min_humedad = humedad_dias[0]
suma = 0

for h in humedad_dias:
    if h > max_humedad:
        max_humedad = h
    if h < min_humedad:
        min_humedad = h
    suma += h

promedio_humedad = suma / len(humedad_dias)

print("Informe semanal de humedad:")
print("Máximo:", max_humedad)
print("Mínimo:", min_humedad)
print("Promedio:", round(promedio_humedad, 2))
```

```python
#Ejercicio 5: Consola de Registro de Datos Manual
#Implementar un menú interactivo por consola donde el usuario pueda:
#Agregar nuevas mediciones (pidiendo dato y día).
#Consultar todos los datos ingresados.
#Calcular estadísticas simples (promedio, máximo, mínimo).
#Esto integra bucles, condicionales, manejo de listas, y práctica de interacción con input/output.

datos = []

def mostrar_menu():
    print("\nMenú de registro de datos")
    print("1. Agregar medición")
    print("2. Consultar datos")
    print("3. Estadísticas")
    print("4. Salir")

def agregar_dato():
    dia = input("Ingrese el día (ejemplo: Lunes): ")
    try:
        valor = float(input("Ingrese el valor de la medición: "))
        datos.append((dia, valor))
        print("Dato agregado.")
    except:
        print("Valor no válido.")

def consultar_datos():
    if len(datos) == 0:
        print("No hay datos registrados.")
    else:
        for d in datos:
            print(f"Día: {d[0]}, Valor: {d[1]}")

def estadisticas():
    if len(datos) == 0:
        print("No hay datos para calcular estadísticas.")
        return
    valores = [d[1] for d in datos]
    max_valor = valores[0]
    min_valor = valores[0]
    suma = 0
    for v in valores:
        if v > max_valor:
            max_valor = v
        if v < min_valor:
            min_valor = v
        suma += v
    promedio = suma / len(valores)
    print(f"Máximo: {max_valor}")
    print(f"Mínimo: {min_valor}")
    print(f"Promedio: {round(promedio,2)}")

while True:
    mostrar_menu()
    opcion = input("Seleccione una opción: ")
    if opcion == "1":
        agregar_dato()
    elif opcion == "2":
        consultar_datos()
    elif opcion == "3":
        estadisticas()
    elif opcion == "4":
        print("Saliendo...")
        break
    else:
        print("Opción no válida.")
```

```python
# Función factorial retorna el factorial del parámtro n
def factorial(n):
  f = 1
  for i in range(1, n + 1):
    f = f * i
  return f

def combinacion(n, r):
  return factorial(n) // (factorial(r) * factorial(n - r))
# Se piden los datos al usuario
n = int(input('Entre valor de n '))
r = int(input('Entre valor de r '))

valor_combinacion = combinacion(n, r)

print(f'Las combinaciones de {n} tomadas de {r} en {r} son: {valor_combinacion}')
```
